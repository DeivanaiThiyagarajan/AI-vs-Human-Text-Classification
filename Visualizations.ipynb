```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('DataSets/Processed_Data.csv',delimiter = ',',encoding='utf-8')
```

```python
data.head()
```

```python
null_rows = data.isnull().sum(axis=1)

# Filter for rows with at least one null value
rows_with_nulls = data[null_rows > 0]

rows_with_nulls.info()
```

```python
value_counts = data['generated'].value_counts()

# Create bar chart
value_counts.plot(kind='bar',color=['blue', 'green'])
plt.title('Labels Value Counts')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='generated', y='avg_len_sentences', data=data, palette='Set2')
plt.title("Boxplot of Average Sentence Length by Generated Label", fontsize=16)
plt.xlabel("Generated Label (0: Human, 1: AI)", fontsize=12)
plt.ylabel("Average Sentence Length", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='generated', y='punctuations_count', data=data, palette='Set3')
plt.title("Boxplot of Punctuation Count by Generated Label", fontsize=16)
plt.xlabel("Generated Label (0: Human, 1: AI)", fontsize=12)
plt.ylabel("Punctuation Count", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='generated', y='words_count', data=data, palette='Set3')
plt.title("Boxplot of Word Count by Generated Label", fontsize=16)
plt.xlabel("Generated Label (0: Human, 1: AI)", fontsize=12)
plt.ylabel("Punctuation Count", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='generated', y='readability_score', data=data, palette='Set2')
plt.title("Boxplot of readability score by Generated Label", fontsize=16)
plt.xlabel("Generated Label (0: Human, 1: AI)", fontsize=12)
plt.ylabel("Average Sentence Length", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='generated', y='stop_word_ratio', data=data, palette='Set2')
plt.title("Boxplot of Stop word ratio by Generated Label", fontsize=16)
plt.xlabel("Generated Label (0: Human, 1: AI)", fontsize=12)
plt.ylabel("Average Sentence Length", fontsize=12)
plt.show()
```
